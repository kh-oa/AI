# Bài 2: Code minimax không giống với slide và nhận xét 2 kết quả.

```python
#Code minimax không giống với slide

from tkinter import *
from tkinter import messagebox
import math
import time

class TicTacToe4x4AI:
    def __init__(self, master):
        self.master = master
        master.title('Game XO 4x4 (Người chơi vs AI)')
        
        self.BOARD_SIZE = 4
        self.HUMAN_PLAYER = 'X'
        self.AI_PLAYER = 'O'
        self.WINNING_SCORE = 10
        self.LOSING_SCORE = -10
        self.MAX_MINIMAX_DEPTH = 4 

        self.buttons = []
        self.board = [['' for _ in range(self.BOARD_SIZE)] for _ in range(self.BOARD_SIZE)]
        self.current_turn = self.HUMAN_PLAYER
        self.game_over = False
        self.moves_made = 0

        self.create_buttons()
        self.create_menu()
        self.reset_game()

    def create_buttons(self):
        for r in range(self.BOARD_SIZE):
            row_buttons = []
            for c in range(self.BOARD_SIZE):
                button = Button(self.master, text="", font=("Helvetica", 20), height=3, width=7,
                                 bg="SystemButtonFace", command=lambda row=r, col=c: self.player_move(row, col))
                button.grid(row=r, column=c)
                row_buttons.append(button)
            self.buttons.append(row_buttons)

    def create_menu(self):
        game_menu = Menu(self.master)
        self.master.config(menu=game_menu)

        option_menu = Menu(game_menu, tearoff=False)
        game_menu.add_cascade(label="Options", menu=option_menu)
        option_menu.add_command(label="Restart Game", command=self.reset_game)

    def disable_all_buttons(self):
        for r in range(self.BOARD_SIZE):
            for c in range(self.BOARD_SIZE):
                self.buttons[r][c].config(state=DISABLED)

    def enable_all_buttons(self):
        for r in range(self.BOARD_SIZE):
            for c in range(self.BOARD_SIZE):
                self.buttons[r][c].config(state=NORMAL)
                self.buttons[r][c].config(bg="SystemButtonFace")
                self.buttons[r][c]["text"] = ""
                self.board[r][c] = ''

    def check_win(self, current_board, player):
        for i in range(self.BOARD_SIZE):
            if all(current_board[i][j] == player for j in range(self.BOARD_SIZE)):
                return True
            if all(current_board[j][i] == player for j in range(self.BOARD_SIZE)):
                return True
        
        if all(current_board[i][i] == player for i in range(self.BOARD_SIZE)):
            return True
        
        if all(current_board[i][self.BOARD_SIZE-1-i] == player for i in range(self.BOARD_SIZE)):
            return True
            
        return False

    def is_board_full(self, current_board):
        return all(current_board[r][c] != '' for r in range(self.BOARD_SIZE) for c in range(self.BOARD_SIZE))

    def get_empty_cells(self, current_board):
        empty_cells = []
        for r in range(self.BOARD_SIZE):
            for c in range(self.BOARD_SIZE):
                if current_board[r][c] == '':
                    empty_cells.append((r, c))
        return empty_cells

    def evaluate(self, current_board):
        if self.check_win(current_board, self.AI_PLAYER):
            return self.WINNING_SCORE
        elif self.check_win(current_board, self.HUMAN_PLAYER):
            return self.LOSING_SCORE
        else:
            return 0

    def minimax(self, current_board, depth, is_maximizing_player, alpha, beta):
        score = self.evaluate(current_board)

        if score == self.WINNING_SCORE or score == self.LOSING_SCORE or \
           self.is_board_full(current_board) or depth == self.MAX_MINIMAX_DEPTH:
            return score

        if is_maximizing_player:
            best_val = -math.inf
            for r, c in self.get_empty_cells(current_board):
                current_board[r][c] = self.AI_PLAYER
                value = self.minimax(current_board, depth + 1, False, alpha, beta)
                current_board[r][c] = ''
                best_val = max(best_val, value)
                alpha = max(alpha, best_val)
                if beta <= alpha:
                    break
            return best_val
        else:
            best_val = math.inf
            for r, c in self.get_empty_cells(current_board):
                current_board[r][c] = self.HUMAN_PLAYER
                value = self.minimax(current_board, depth + 1, True, alpha, beta)
                current_board[r][c] = ''
                best_val = min(best_val, value)
                beta = min(beta, best_val)
                if beta <= alpha:
                    break
            return best_val

    def find_best_move(self):
        best_val = -math.inf
        best_move = (-1, -1)

        for r, c in self.get_empty_cells(self.board):
            self.board[r][c] = self.AI_PLAYER
            move_val = self.minimax(self.board, 0, False, -math.inf, math.inf)
            self.board[r][c] = ''

            if move_val > best_val:
                best_val = move_val
                best_move = (r, c)
        return best_move

    def handle_game_end(self):
        if self.check_win(self.board, self.HUMAN_PLAYER):
            messagebox.showinfo("OX Game", "Bạn (Player X) thắng cuộc!!")
            self.game_over = True
        elif self.check_win(self.board, self.AI_PLAYER):
            messagebox.showinfo("OX Game", "AI (Player O) thắng cuộc!!")
            self.game_over = True
        elif self.is_board_full(self.board):
            messagebox.showinfo("OX Game", "HÒA!!")
            self.game_over = True
        
        if self.game_over:
            self.disable_all_buttons()

    def player_move(self, row, col):
        if self.game_over or self.current_turn != self.HUMAN_PLAYER or self.board[row][col] != '':
            messagebox.showwarning("OX Game", "LỖI!! Vui lòng chọn lại ô trống hoặc không phải lượt của bạn.")
            return

        self.board[row][col] = self.HUMAN_PLAYER
        self.buttons[row][col]["text"] = self.HUMAN_PLAYER
        self.moves_made += 1

        self.handle_game_end()
        if self.game_over:
            return

        self.current_turn = self.AI_PLAYER
        self.master.after(200, self.ai_move) 

    def ai_move(self):
        if self.game_over:
            return

        best_move_r, best_move_c = self.find_best_move()

        if best_move_r != -1 and best_move_c != -1:
            self.board[best_move_r][best_move_c] = self.AI_PLAYER
            self.buttons[best_move_r][best_move_c]["text"] = self.AI_PLAYER
            self.buttons[best_move_r][best_move_c].config(bg="#FFA500") 
            self.moves_made += 1
            self.master.update_idletasks()

            self.handle_game_end()
            if self.game_over:
                return
        
        self.current_turn = self.HUMAN_PLAYER

    def reset_game(self):
        self.enable_all_buttons()
        self.board = [['' for _ in range(self.BOARD_SIZE)] for _ in range(self.BOARD_SIZE)]
        self.current_turn = self.HUMAN_PLAYER
        self.game_over = False
        self.moves_made = 0
        messagebox.showinfo("OX Game", "Game đã được khởi động lại. Bạn (X) đi trước!")

if __name__ == "__main__":
    root = Tk()
    game = TicTacToe4x4AI(root)
    root.mainloop()
```
